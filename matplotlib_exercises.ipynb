# 1
Use matplotlib to plot the following equation:

```python
import matplotlib.pyplot as plt
import numpy as np
import math
from random import randint
import pandas as pdimport 
```

```python
# create range for x
x = range(-50, 50)

# calculate the y value for each element of the x vector
y = [n **2 + 2*n + 2 for n in x]
plt.title("$X**2 + 2*n + 2$")
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.annotate('(0, 0) The origen', 
             xy = (0, 0), 
             xytext =(0, 1000),
            arrowprops = dict(facecolor = 'gray'))
plt.plot(x, y, 
        color = 'k',
         linestyle = '--')


```

# 2
Create and label 4 separate charts for the following equations (choose a range for x that makes sense):

```python
# y = √x
plt.title("y = √x")
x1 = range(1, 50)
a = [math.sqrt(n) for n in x1]
plt.plot(x1, a)
plt.grid(True);
```

```python
# y = x**3
x2 = range(0, 35)
plt.title("y = x**3")
b = [n**3 for n in x2]
plt.plot(x2, b)
plt.grid(True);
```

```python
# y = 2**x
x3 = [x * 0.1 for x in range(0,150)]
plt.title("y = 2**x")
c = [2**n for n in x3]
plt.grid(True);
plt.plot(x3, c)
```

```python
# y = 1/(x+1)
x4 = range(0, 50)
plt.title("y = 1/(x+1)")
d = [1/(n+1) for n in x4]

plt.plot(x4, d)
plt.grid(True);
```

# 3
Combine the figures you created in the last step into one large figure with 4 subplots

```python
plt.figure(figsize = (10,8))
fig, axs = plt.subplots(2, 2)
axs[0, 0].plot(x1, a)
axs[0, 0].set_title("Y = √x")
axs[0, 0].grid(True)
axs[1, 0].plot(x2, b)
axs[1, 0].set_title("Y = x**3")
axs[1, 0].grid(True)
axs[0, 1].plot(x3, c)
axs[0, 1].set_title(" Y = 2x")
axs[0, 1].grid(True)
axs[1, 1].plot(x4 , d)
axs[1, 1].set_title("Y = 1/(x+1)")
axs[1, 1].grid(True)
fig.tight_layout()
```

# 4
Combine the figures you created in the last step into one figure where each of the 4 equations has a different color for the points. Be sure to include a legend and an appropriate title for the figure

```python
plt.figure(figsize = (10,8))
plt.plot(x1, a, label = "Y = √x")
plt.plot(x2, b, label = "Y = x**3")
plt.plot(x3, c, label = " Y = 2x")
plt.plot(x4, d, label = "Y = 1/(x+1)")
plt.xlabel('X label')
plt.ylabel('"ABCD" label')
plt.title("EXERCISE 4")
plt.ylim(0, 10)
plt.xlim(0, 10)
plt.legend()
plt.show()
```
